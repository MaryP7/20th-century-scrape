**In the 20th century scrape text file is a bit messy.punctation marks, long spacaes between lines and the text stcture should be fixed in a computer readable format. So, text wrangling is requuired.**

#text wrangling steps

```python
# Split the text into sentences
sentences = nltk.sent_tokenize(text)
```

```python
# replace '\n' & '\t' from text
sentences = [line.replace('\n', ' ').replace('\t', ' ') for line in sentences]
```

```python
print(sentences)
```

```python
# remove more than one spaces  before lines in text

sentences = [re.sub(' +', ' ', line) for line in sentences]

print(sentences)
```

```python
# Remove punctuation
sentences = [line.replace(",", "").replace("!", "").replace("?", "").replace(":", "").replace(";", "").replace('"', '').replace("'", "") for line in sentences]

print(sentences)
```

```python
# Remove non-alphanumeric characters

def remove_special_characters(line):
    line = re.sub(r'[^a-zA-Z0-9\s]', '', line)
    return line

sentences = [remove_special_characters(line) for line in sentences]
```

```python
# remove spceial characters except . from lines
sentences = [re.sub(r'[^a-zA-Z0-9.\s]', '', line) for line in sentences]

```

```python
print(sentences)
```

```python
# Join the sentences back into a string
text = '\n'.join(sentences)
print (text)
```

**Named Entity Recognition (NER) object is a tool or model used to identify and classify named entities in text. These entities can be names of persons, organizations, locations, dates, times, quantities, monetary values, percentages, and more.**

```python
# Use the text to create a NER object.
# Split the sentence entities from the NER object.

import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

# Split the sentence entities from the NER object
sentence_entities = []
for sentence in doc.sents:
    entities = [(ent.text, ent.label_) for ent in sentence.ents]
    sentence_entities.append(entities)

print(sentence_entities)
```

```python
# Creating the relationships dataframe.

import pandas as pd
# Creating a dictionary to store the relationships
relationships = {}

# Iterate over each sentence
for sentence in doc.sents:
    # Extract the entities from the sentence
    entities = [(ent.text, ent.label_) for ent in sentence.ents]

    # Iterate over each pair of entities in the sentence
    for i in range(len(entities)):
        for j in range(i + 1, len(entities)):
            key = (entities[i][0], entities[j][0])
            if key not in relationships:
                relationships[key] = 1
            else:
                relationships[key] += 1

# Create a Pandas DataFrame from the dictionary
relationships_df = pd.DataFrame(relationships, index=[0]).T
relationships_df.columns = ['Count']

# Print the DataFrame
relationships_df
```
